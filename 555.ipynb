```python
import numpy as np

# Input feature (X) and label (y)
X = np.array([1, 2, 3, 4])
y = np.array([0, 0, 1, 1])

# Initialize weight and bias
w = 0.0
b = 0.0
lr = 0.1

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# One training step
z = w * X + b
y_pred = sigmoid(z)

# Gradient calculation
dw = np.mean((y_pred - y) * X)
db = np.mean(y_pred - y)

# Update
w -= lr * dw
b -= lr * db

print("Predicted Probabilities:", y_pred)
print("Updated weight:", w)
print("Updated bias:", b)
```
